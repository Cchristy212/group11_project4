```python
import pandas as pd
```

```python
# read in the csv's
year_13 = pd.read_csv('2013_2021Brakets/cbb13.csv')
year_14 = pd.read_csv('2013_2021Brakets/cbb14.csv')
year_15 = pd.read_csv('2013_2021Brakets/cbb15.csv')
year_16 = pd.read_csv('2013_2021Brakets/cbb16.csv')
year_17 = pd.read_csv('2013_2021Brakets/cbb17.csv')
year_18 = pd.read_csv('2013_2021Brakets/cbb18.csv')
year_19 = pd.read_csv('2013_2021Brakets/cbb19.csv')
year_21 = pd.read_csv('2013_2021Brakets/cbb21.csv')
```

```python
# Add the year column
year_13['YEAR'] = 2013
year_14['YEAR'] = 2014
year_15['YEAR'] = 2015
year_16['YEAR'] = 2016
year_17['YEAR'] = 2017
year_18['YEAR'] = 2018
year_19['YEAR'] = 2019
year_21['YEAR'] = 2021
```

```python
# Select the teams that made it to the tournament
year_13 = year_13.dropna(subset=['SEED'])
year_14 = year_14.dropna(subset=['SEED'])
year_15 = year_15.dropna(subset=['SEED'])
year_16 = year_16.dropna(subset=['SEED'])
year_17 = year_17.dropna(subset=['SEED'])
year_18 = year_18.dropna(subset=['SEED'])
year_19 = year_19.dropna(subset=['SEED'])
year_21 = year_21.dropna(subset=['SEED'])
```

```python
# Merge the teams that maade the tournament

combined_df = pd.merge(year_13,year_14, how='outer')
combined_df = pd.merge(combined_df,year_15, how='outer')
combined_df = pd.merge(combined_df,year_16, how='outer')
combined_df = pd.merge(combined_df,year_17, how='outer')
combined_df = pd.merge(combined_df,year_18, how='outer')
combined_df = pd.merge(combined_df,year_19, how='outer')
combined_df = pd.merge(combined_df,year_21, how='outer')
```
