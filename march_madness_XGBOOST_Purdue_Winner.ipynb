## Import Dependencies

```python
import pandas as pd
import numpy as np
```

## Load Data

```python
# read in the csv's
year_13 = pd.read_csv('Data/cbb13.csv')
year_14 = pd.read_csv('Data/cbb14.csv')
year_15 = pd.read_csv('Data/cbb15.csv')
year_16 = pd.read_csv('Data/cbb16.csv')
year_17 = pd.read_csv('Data/cbb17.csv')
year_18 = pd.read_csv('Data/cbb18.csv')
year_19 = pd.read_csv('Data/cbb19.csv')
year_21 = pd.read_csv('Data/cbb21.csv')
```

```python
# Add the year column
year_13['YEAR'] = 2013
year_14['YEAR'] = 2014
year_15['YEAR'] = 2015
year_16['YEAR'] = 2016
year_17['YEAR'] = 2017
year_18['YEAR'] = 2018
year_19['YEAR'] = 2019
year_21['YEAR'] = 2021
```

## Perform Data Cleaning

```python
# Drop the teams that did not made it to the tournament each year (seed is NaN)
year_13 = year_13.dropna(subset=['SEED'])
year_14 = year_14.dropna(subset=['SEED'])
year_15 = year_15.dropna(subset=['SEED'])
year_16 = year_16.dropna(subset=['SEED'])
year_17 = year_17.dropna(subset=['SEED'])
year_18 = year_18.dropna(subset=['SEED'])
year_19 = year_19.dropna(subset=['SEED'])
year_21 = year_21.dropna(subset=['SEED'])
```

```python
# Merge the dataframes for each year for teams that made the tournament

combined_df = pd.merge(year_13,year_14, how='outer')
combined_df = pd.merge(combined_df,year_15, how='outer')
combined_df = pd.merge(combined_df,year_16, how='outer')
combined_df = pd.merge(combined_df,year_17, how='outer')
combined_df = pd.merge(combined_df,year_18, how='outer')
combined_df = pd.merge(combined_df,year_19, how='outer')
combined_df = pd.merge(combined_df,year_21, how='outer')
```

```python
combined_df
```

```python
# Drop the CONF, G, and POSTSEASON columns as these are not needed for our analysis
combined_df = combined_df.drop(columns=['CONF','G','POSTSEASON'])
```

```python
combined_df.head()
```

```python
# Read in the tournament results csv file
tourney_results = pd.read_csv('Data/data_cleaned_updated.csv')
```

```python
tourney_results.head()
```

```python
# Keep only the results of games from 2013 to 2021
new_tourney_results_df = tourney_results.iloc[0:535]
```

```python
new_tourney_results_df
```

```python
# Drop ROUND column as this is not needed for our analysis
new_tourney_results_df = new_tourney_results_df.drop(columns=['ROUND'])
```

```python
new_tourney_results_df.head()
```

```python
# Merge the two datasets and add a prefix to the stats columns to indicate which team they're for
total_df = new_tourney_results_df.merge(combined_df.add_prefix('Team1'), how='left', left_on=['TEAM', 'YEAR'], 
                                        right_on=['Team1TEAM', 'Team1YEAR']).drop(['Team1TEAM'], axis=1).merge(combined_df.add_prefix('Team2'), 
                                        how='left', left_on=['TEAM', 'YEAR'], right_on=['Team2TEAM', 'Team2YEAR']).drop(['Team2TEAM'],axis=1)
```

```python
total_df
```

```python
# Add a new column that will contain '1' if Team1 won the game and a '0' if Team1 lost
outcomes = [total_df['SCORE'] > total_df['SCORE.1'],
           total_df['SCORE'] < total_df['SCORE.1']]

results = [1,0]

# This column will be our target for the learning model
total_df['Team1Wins'] = np.select(outcomes, results, 1)
```

```python
total_df.head()
```

# Filtering Data for UCLA

```python
# Drop YEAR, SCORE, TEAM, Team1SEED, TEAM.1, SCORE.1, Team2SEED, Team1YEAR, Team2YEAR as these aren't neeeded for our analysis
total_df = total_df.drop(columns=['YEAR', 'SCORE', 'SCORE.1', 'TEAM','Team1SEED', 'TEAM.1', 'Team2SEED', 'Team1YEAR', 'Team2YEAR'])

# Rename the SEED and SEED.1 columns
total_df.rename(columns={'SEED': 'Team1Seed', 'SEED.1': 'Team2Seed'})

total_df.head()
```

```python
# Define the X (features) and y (target) sets
y = total_df["Team1Wins"].values
X = total_df.drop("Team1Wins", axis=1)
```

```python
# Import Train Test Split
from sklearn.model_selection import train_test_split
```

```python
# Divide data into training and testing sets

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

## Create Machine Learning Model - Logistic Regression

## Create Machine Learning Model - XGBoost

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
march_madness_classifier = XGBClassifier()
march_madness_classifier.fit(X_train, y_train)
xgbPredictions = march_madness_classifier.predict(X_test)
print(f"Training Data Score: {march_madness_classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {march_madness_classifier.score(X_test, y_test)}")
```
